```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "/workspaces/final-project-karina-munoz/data/raw/Pantheon+SH0ES.dat",
    delim_whitespace=True,
    comment="#",         # ignores header comments
    skip_blank_lines=True)

cols = [                    #pulling only necessary columns
    "CID",  #Supernova ID
    "RA",   #locations
    "DEC",
    "zHD",  # redshift
    "zCMB", #redshift relative to cosmic microwave background
    "MU_SH0ES", #distance modulus
    "MU_SH0ES_ERR_DIAG" #uncertainties 
]
df2 = df[cols].copy()
df2.info()
```

```python
# d_L_Mpc distance in megaparsecs
u = df2['MU_SH0ES']
d_L_Mpc = 10 ** (((u+5)/5)-6)   #distance-modulus eqn
df2["d_L_Mpc"] = d_L_Mpc
df2.info()
```

```python
# recession velocity

c = 299792  # speed of light in km/s
z = df2["zHD"]

numer = ((1+z)**2) - 1        #velocity derived from redshift z 
denom = ((1+z)**2) + 1
velocity = c * numer/denom

df2["velocity_kms"] = velocity
df2.info()
```

```python
df2.to_csv("/workspaces/final-project-karina-munoz/data/processed/clean_sn_data.csv",
           index=False)
```
