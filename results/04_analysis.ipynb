At low redshifts (z < 0.1), the expansion of the universe is well-approximated by a linear Hubble relation: v = H0d

My linear model found:

* H0 ≈ 64.63 km/s/Mpc
* intercept ≈ 539 km/s (expected due to measurement noise + local motions)
* R² ≈ 0.983, indicating the linear model explains most of the variance

Supernovae in the nearby universe follow a nearly perfect linear expansion trend, consistent with classical expectations of Hubble’s Law.

Before isolating the low-z region, the luminosity distance vs. recession velocity plot showed strong curvature. This curvature arises because luminosity distance is a nonlinear cosmological quantity d_L = 1(+z)r(z) 

At moderate or high redshift, the accelerated expansion of the universe makes the relation deviate from linearity.
Even after selecting z < 0.1, my residual analysis revealed:
* Slight curvature at larger distances
* Deviations consistent with higher-order cosmological effects
* Evidence that the purely linear model begins to break down > 300 Mpc

The quadratic model captured this curvature slightly better, but this improvement is small and likely reflects cosmological distance-measurement systematics rather than true acceleration in the low-z regime.

| --------- | ----- | -------- | -------- |
| Linear    | 0.983 | 711 km/s | 492 km/s |
| Quadratic | 0.986 | 651 km/s | 448 km/s |
